## Exploratory Data Analysis

### Qual e o Percentual de Acidentes com Vitimas?
**<font color='red'>Cerca de 35% dos Acidentes envolve pelo menos uma Vitima</font>**

```python
df['TipoDeOcorrencia'] = df['TipoDeOcorrencia'].apply(lambda x: str(x).upper())
df['Vitimas'] = df['TipoDeOcorrencia'].apply(lambda x: 1 if ('COM VÍTIMA' in x or 'COM VITIMA' in x
                                                             or 'COM MORTE' in x or 'FATAL' in x )else 0)
```

```python
df['Vitimas'].value_counts( normalize=True)
```

```python
plt.rcdefaults()
f, ax = plt.subplots(1,2, figsize=(10,5))
df['Vitimas'].value_counts().plot.pie(explode=[0,0.1], autopct='%1.1f%%', ax=ax[0], shadow=True)
ax[0].set_title('Vitimas')
ax[0].set_ylabel('')
sns.countplot('Vitimas', data= df, ax=ax[1])
ax[1].set_title('Vitimas')
plt.show()
```

###  Quais os tipos de Acidentes com maior possibilidade de ter pelo menos uma vitima Fatal?

### Quais os veiculos que mais se envolveram em Acidentes Fatais?
**<font color='red'>Automovel - 60%, Caminhão 58% e Moto 42%</font>**

```python
df['Automovel'] = df['Automovel'].apply(lambda x: float(0) if(str(x).strip() == '') else float(x))
df['Caminhao'] = df['Caminhao'].apply(lambda x: float(0) if(str(x).strip() == '') else float(x))
df['Moto'] = df['Moto'].apply(lambda x: float(0) if(str(x).strip() == '') else float(x))
df['Transp__Cargas_Especiais'] = df['Transp__Cargas_Especiais'].apply(lambda x: float(0) if(str(x).strip() == '') else float(x))
df['Utilitario'] = df['Utilitario'].apply(lambda x: float(0) if(str(x).strip() == '') else float(x))
df['Automovel'].value_counts()
```

```python
df.plot.density(y='Automovel', xlim=(0,15), figsize=(6,3))
```

```python
df_fatal = df[df['Fatal'] ==1]
```

```python
veiculos = ['Automovel', 'Bicicleta', 'Caminhao', 'Moto', 'Onibus', 'Outros',
       'Tracao_animal', 'Transp__Cargas_Especiais', 'Trator__maquinas',
       'Utilitario']
df_veiculos = []
for i in  range(len(veiculos)):
    percentual = df_fatal[veiculos[i]].count() /df_fatal.shape[0]
    df_veiculos.append([veiculos[i], percentual])
df2 = pd.DataFrame(np.array(df_veiculos), columns=['Veiculo', 'Percentual'])
df2
```

```python
plt.figure(figsize=(8,5))
df2['Percentual'] = df2['Percentual'].astype(float)
sns.barplot(x='Veiculo', y = 'Percentual', data= df2)
plt.xticks(rotation=90);
```

### Evolução de Acidentes com Vitimas Fatais ao Longo dos Anos
**<font color='red'>Houve uma queda de 14% no numero de acidentes com vitimas fatais de 2015 a 2020</font>**

```python
df['Data'] = pd.to_datetime(df['Data'] )
df['year'] = df['Data'].dt.year
df = df[df['year']<2021]
df_fatal = df[df['Fatal'] ==1]
df_vitimas = df[df['Vitimas'] ==1]
```

```python
#df.groupby(df['Data'].dt.to_period('M'))['No_DaOcorrencia'].count().plot.bar()
#df = df[df['year']<2021]
fatal = df.groupby(df_fatal['year'])['No_DaOcorrencia'].count()
vitimas = df.groupby(df_vitimas['year'])['No_DaOcorrencia'].count()
fatal
```

```python
f, ax = plt.subplots(1,2, figsize=(10,4))
fatal.plot.line(ax=ax[0])
ax[0].set_title('Acidentes com vitimas fatais')
ax[0].set_xlabel('Ano')
vitimas.plot.line(ax=ax[1])
ax[1].set_xlabel('Ano')
ax[1].set_title('Acidentes com vitimas');
```

```python
result= pd.merge(fatal.to_frame(), vitimas.to_frame(),on="year" )
result = result.rename(columns={'No_DaOcorrencia_x': 'Fatais'})
result = result.rename(columns={'No_DaOcorrencia_y': 'Vitimas'})
```
